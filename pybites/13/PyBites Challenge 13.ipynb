Given a CSV which contains movies statistics:
    
1. We are only interested in those whose director has 4 or more movies
1. For each director, compute the average IMDB score
1. Display movies groupped by directors along with average IMDB score

### First Step: Read the CSV and Filter In Relevant Columns

```python
import pandas as pd

DIRECTOR = 'director_name'
TITLE = 'movie_title'
SCORE = 'imdb_score'
RELEVANT_COLUMNS = [DIRECTOR, TITLE, SCORE]
COUNT_THRESHOLD = 4

df = pd.read_csv('movie_metadata.csv')
df = df.filter(items=RELEVANT_COLUMNS)
df
```

### Filter Out Entries Whose Directors Have Less Than 4 Titles

```python
df = df.groupby(DIRECTOR).filter(lambda row: len(row) >= COUNT_THRESHOLD)
for director, group in df.groupby(DIRECTOR):
    print()
    print(director)
    print(group)
```
