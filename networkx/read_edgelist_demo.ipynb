```python
import networkx as nx
import matplotlib.pyplot as plt
import warnings

# Ignore deprecated warnings and others generated by networkx
warnings.filterwarnings(action='ignore', category=DeprecationWarning)
warnings.filterwarnings(action='ignore', category=UserWarning)

# Create and edge-list file
data_filename = 'read_edgelist_demo.txt'
with open(data_filename, 'w') as f:
    f.write('greeting.h myapp.cpp\n')
    f.write('database.h myapp.cpp\n')
    f.write('hello.h greeting.h\n')
    f.write('goodbye.h greeting.h\n')    
```

```python
g = nx.read_edgelist(data_filename, create_using=nx.DiGraph())

nx.draw(
    g,
    with_labels=True,
    font_color='white',
    node_size=5000,
)
plt.show()
```

```python
print('Build Order:')
for node in nx.topological_sort(g):
    print('-', node)
```
